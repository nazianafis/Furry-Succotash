```python
import random
import math
import matplotlib.pyplot as plt

from samila import GenerativeImage, Projection
```

```python
DEFAULT_SEED = 42
#1018273
```

```python
def f1(x,y):
    result = random.uniform(-1,1) * x  - math.sin(y**2)
    return result

def f2(x,y):
    result = random.uniform(-1,1) * y - math.cos(x**2) + 2**x
    return result

g = GenerativeImage(f1,f2)
g.generate(seed=DEFAULT_SEED)
g.seed

g.plot(size=(8,8), projection=Projection.POLAR, color = 'black', bgcolor = 'white')
```

```python
def f1(x,y):
    result = random.uniform(-1,1) * x**2  + math.sin(y**2) + abs(y-x)
    return result

def f2(x,y):
    result = random.uniform(-1,1) * y**2 - math.cos(x**2) + 2*x
    return result

g = GenerativeImage(f1,f2)
g.generate(seed=DEFAULT_SEED)
g.seed

g.plot(size=(8,8), projection=Projection.POLAR, color = 'black', bgcolor = 'white')
```
